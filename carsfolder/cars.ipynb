```python
import pandas as pd
cars = pd.read_csv('cars93.csv')
```

```python
cars.columns
```

```python
car_manufacturer_sizes= cars[['Manufacturer','Type']]
car_sizes = car_manufacturer_sizes.Type.value_counts()
car_sizes.plot.bar()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
 #keeps plots in the notebook
```

```python
ax = car_sizes.plot(title = 'Distribution of Type in Cars93 Dataset',kind='bar')
ax.set_xlabel('Type')
ax.set_ylabel('Number of cars')
plt.show()
```

```python
ax = car_sizes.plot(title = 'Distribution of Type in Cars93 Dataset',kind='barh')
ax.set_xlabel('Number of cars')
ax.set_ylabel('Type')
plt.show()
```

```python
ax = cars.plot(title = 'Weight of Vehicles vs MPG', kind='scatter', x='Weight',y='MPG.city')
ax.set_xlabel('Weight')
ax.set_ylabel('MPG (City)')
plt.show()
```

```python
ax = cars.plot(title = 'Weight of Vehicles vs MPG', kind='scatter', x='Weight',y='MPG.highway')
ax.set_xlabel('Weight')
ax.set_ylabel('MPG (Highway)')
plt.show()
```

```python
plt.scatter(cars.Weight,cars['MPG.city'],c='#838286',label='MPG (City)')
plt.scatter(cars.Weight,cars['MPG.highway'],c='#ed1650', label='MPG (Highway)')
plt.title('Weight of Vehicles vs MPG')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.legend(loc=3)
plt.show()
```
