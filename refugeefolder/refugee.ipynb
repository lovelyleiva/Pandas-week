```python
refasylumT = refasylum.copy()
refasylumT = refasylumT.drop("Country_Code", axis = 1)
refasylumT = refasylumT.T
refasylumT.columns = refasylumT.iloc[0]
refasylumT = refasylumT.rename(columns = {"Country_Name":"Year"})
refasylumT.reset_index(level=0, inplace =True)
refasylumT = refasylumT.rename(columns={"index":"Year"})
refasylumT = refasylumT.drop([0])
print(refasylumT.columns)
refasylumT.head(10)
```
