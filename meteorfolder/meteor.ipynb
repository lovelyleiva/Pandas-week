```python
import pandas as pd
meteor = pd.read_csv('Meteorite_Landings.csv')
```

```python
meteor
```

```python
#remove the nametype field
meteor.drop('nametype',inplace=True,axis=1) 
#inplace True so it does it in meteor and keeps the change
#set axis to 0 for rows or 1 for columns
meteor
```

```python
meteor.columns = [c.replace(' (g)','_g') for c in meteor.columns]
```

```python
#clean mass(g) field so that there's default of 0 where there is no mass listed
meteor['mass_g'].fillna('0', inplace = True)
```

```python
#story each decade's data in it's own data sheet in Excel using pandas
#so 1990-1999,2000-2009, 2010-2019, etc
meteor.groupby('year')['name'].apply(lambda x: ",".join(x))
```

```python
#should do a list comprehension here.
bins = [0, 1000, 1500, 1600, 1700, 1800]
recentdecades = [num for num in range(1900,2040) if num % 10 == 0]

for i in recentdecades:
    bins.append(i)

labels = ['0-999','1000-1499','1500-1599','1600-1699','1700-1799','1800-1899','1900-1909','1910-1919','1920-1929','1930-1939','1940-1949','1950-1959','1960-1969','1970-1979','1980-1989','1990-1999','2000-2009','2010-2019','2020-']
meteor['binned']=pd.cut(meteor['year'], bins=bins, labels=labels)
meteor.head()
```

```python
meteordecades= meteor.groupby('binned')['name'].apply(lambda x: ','.join(x))
meteordecades
```
