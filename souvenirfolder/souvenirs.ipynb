```python
# Using souvenir.csv as the base:

# Read in all of the other CSVs
# Replace the ID fields with the descriptions
# Save the new file to souvenirs.JSON
# Create separate CSV files for each owner's set of souvenirs.
```

```python
# Read in all of the other CSVs

import pandas as pd
souvenir = pd.read_csv('souvenir.csv')
city = pd.read_csv('city.csv')
category = pd.read_csv('category.csv')
country = pd.read_csv('country.csv')
location = pd.read_csv('location.csv')
region = pd.read_csv('region.csv')
owner = pd.read_csv('owner.csv')
```

```python
# Replace the ID fields with the descriptions
Master1 = souvenir.merge(city, how = 'outer', on = 'CityID')
Master1.rename(columns={"Name_x":"SouvenirName","Name_y":"CityName"},inplace=True )
Master2 = Master1.merge(category, how ='outer', on = 'CategoryID')
Master2.rename(columns ={"Name":"CategoryName"}, inplace = True)
Master3 = Master2.merge(location, how = "outer", on ="LocationID") 
Master4 = Master3.merge(owner, how = "outer", on = "OwnerID")
Master4.rename(columns={"Name":"Owner"}, inplace=True)
Souvenirs = Master4[['ItemID','ItemDescription','DateObtained','Price','Weight','SouvenirName','CityName','CategoryName','PlaceName','Latitude','Longitude','Owner']]
Souvenirs
#There is no countryID
#There are diff joins, just like SQL
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html
#souvenirs_with_categories.rename(columns={"Name_x":"SouvenirName","Name_y":"CategoryName"},inplace=True)
```

```python
# Save the new file to souvenirs.JSON
souvenirs = Souvenirs.to_json()
print(souvenirs)
# wikijson = wiki_airports.to_json()
# print(wikijson)
```

```python
bills = Souvenirs[['Owner','CategoryName','Price']]

def calc_tips_10percent(x):
    return (x * .1)

bills['tip'] = bills.Price.apply(calc_tips_10percent)
bills['final_total'] = bills['Price'] + bills['tip']
bills.head()
```

```python
bills = Souvenirs[['Owner','CategoryName','Price']]
bills.head()

def calc_tips_percent(total,percentage):
    return (total * percentage/100)

bills['tip'] = bills.Price.apply(calc_tips_percent,percentage=10)
bills['final_total'] = bills['Price'] + bills['tip']
bills.head()
```
